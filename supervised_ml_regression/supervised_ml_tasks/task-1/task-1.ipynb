```python
import pandas as pd 
df = pd.read_csv("data.csv")
df
```

```python
df.drop('date',axis = 1,inplace=True)
df


```

```python
price_col = df.pop("price")
df['price'] = price_col
df
```

```python
df.head()
```

```python
df.columns
```

```python
cat_cols = ['street', 'city', 'statezip', 'country']
for col in cat_cols:
    print(f"----------{col} : {df[col].value_counts()}")
```

```python
# country is redundant then i am dropping it 
df.drop("country",axis = 1,inplace=True)
```

```python
df.head()
```

```python
cat_cols = ["statezip"]
for col in cat_cols:
    print(f"----------{col} : {df[col].value_counts()}")
```

```python
#handling statezip 
statezip_mean_price = df.groupby("statezip")["price"].mean()
df['statezip_target'] = df['statezip'].map(statezip_mean_price)
df

```

```python
df.drop("statezip",axis = 1 , inplace = True)
```

```python
df.head()
```

```python
#now city 
df["city"].nunique()
df["city"].value_counts().head()
```

```python
city_mean_price = df.groupby("city")["price"].mean()
df["city_target"] = df["city"].map(city_mean_price)

df.drop(columns=["city"], inplace=True)
```

```python
df.head()
```

```python
#now street 
df["street"].nunique()
df["street"].value_counts()
```

```python
df.drop("street",axis = 1, inplace=True)
```

```python
df.head()
```

```python
df['yr_renovated'].value_counts()
```

```python
#for yr_renovated which has 0 in it 
df["was_renovated"] = (df["yr_renovated"] > 0).astype(int)
```

```python
ref_year = df['yr_built'].max()
ref_year
```

```python
df['year_since_renovation'] = 0
mask = df['yr_renovated'] > 0
df.loc[mask,"year_since_renovation"] = ref_year - df.loc[mask,"yr_renovated"]
```

```python
df.head()
```

```python
df.drop(columns=["yr_renovated"], inplace=True)
```

```python
df.head()
```

```python
#handling house age 
df["house_age"] = ref_year - df["yr_built"]
df.drop(columns=["yr_built"], inplace=True)
```

```python
df.head()
```

```python

df["sqft_per_floor"] = df["sqft_above"] / df["floors"]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

num_cols = df.drop(columns=["price"]).columns
df[num_cols] = scaler.fit_transform(df[num_cols])
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import accuracy_score,r2_score,recall_score,precision_score,mean_squared_error

df = df[df["price"] > 0]

X = df.iloc[:,0:-1]
y = df.iloc[:,-1]

```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression

lr_model = LinearRegression()
```

```python
lr_model.fit(X_train,y_train)
```

```python
lr_y_pred = lr_model.predict(X_test)
```

```python
lr_r2_score = r2_score(y_test,lr_y_pred)
lr_r2_score
```

```python
lr_mse = mean_squared_error(y_test,lr_y_pred)
lr_mse
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor(
    max_depth=50,
    min_samples_leaf=20,
    random_state=42
)

dt_model.fit(X_train, y_train)
dt_y_pred = dt_model.predict(X_test)

from sklearn.metrics import r2_score, mean_squared_error
import numpy as np

print("R2:", r2_score(y_test, dt_y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, dt_y_pred)))

```

```python
#now trying random forest 
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(
    n_estimators=300,
    max_depth=15,
    min_samples_leaf=10,
    random_state=42,
    n_jobs=-1
)



rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)

print("R2:", r2_score(y_test, rf_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, rf_pred)))
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge = Ridge()

param_grid = {
    "alpha": [0.1, 1, 10, 50, 100,200,300,500]
}

grid = GridSearchCV(
    ridge,
    param_grid,
    scoring="neg_root_mean_squared_error",
    cv=5
)

grid.fit(X_train, y_train)

best_ridge = grid.best_estimator_

ridge_pred = best_ridge.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, ridge_pred))
r2 = r2_score(y_test, ridge_pred)

print("Tuned RF R2:", r2)
print("Tuned RF RMSE:", rmse)
```

```python
import xgboost as xgb

xgb_model = xgb.XGBRegressor(
    n_estimators=100,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    objective="reg:squarederror",
    early_stopping_rounds=50,   
    random_state=42,
    n_jobs=-1
)

xgb_model.fit(
    X_train,
    y_train,
    eval_set=[(X_test, y_test)],
    verbose=False
)
xgb_pred = xgb_model.predict(X_test)

print("XGBoost R2:", r2_score(y_test, xgb_pred))
print("XGBoost RMSE:", np.sqrt(mean_squared_error(y_test, xgb_pred)))

```

```python
import pickle

with open("xgb_model.pkl","wb") as f:
    pickle.dump(xgb_model,f)
```
